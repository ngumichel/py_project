```python
import sys
sys.path.insert(0, "/Users/Myn/Documents/www/py_project")
import numpy as np
import pandas as pd
from pandas import DataFrame
from settings import INSEE_CODE
```

```python
schools = pd.read_csv('../datas/lycees_france.csv', sep=";")
schools
```

```python
schools = DataFrame(schools, columns=["Etablissement", "Année", "Ville", "Code commune", "Académie", "Taux Brut de Réussite Total séries", "Taux Réussite Attendu France Total séries"])
schools
```

```python
for city in INSEE_CODE:
    schools["Code commune"] = np.where(schools["Code commune"].isin(city[1]), city[0], schools["Code commune"])
schools
```

```python
reduced_data = DataFrame(schools, columns=['Code commune', 'Taux Brut de Réussite Total séries', 'Taux Réussite Attendu France Total séries'])
reduced_data
```

```python
grouped_data = reduced_data.groupby(['Code commune'], as_index=False).mean()
grouped_data
```

```python
cities = pd.read_csv('../datas/villes_france.csv', names=['ville_id', 'ville_departement', 'ville_slug', 'ville_nom', 'ville_nom_simple', 'ville_nom_reel', 'ville_nom_soundex', 'ville_nom_metaphone', 'ville_code_postal', 'ville_commune', 'Code commune', 'ville_arrondissement', 'ville_canton', 'ville_amdi', 'ville_population_2010', 'ville_population_1999', 'ville_population_2012', 'ville_densite_2010', 'ville_surface', 'ville_longitude_deg', 'ville_latitude_deg', 'ville_longitude_grd', 'ville_latitude_grd', 'ville_longitude_dms', 'ville_latitude_dms', 'ville_zmin', 'ville_zmax'], header=None, low_memory=False)
cities
```

```python
reduced_cities = DataFrame(cities, columns=['ville_nom', 'ville_population_2012', 'Code commune'])
reduced_cities
```

```python
merged_data = pd.merge(reduced_cities, grouped_data, on="Code commune")
merged_data
```

```python
merged_data['Taux Objectif'] = merged_data['Taux Brut de Réussite Total séries'] - merged_data['Taux Réussite Attendu France Total séries']
merged_data
```

```python
sorted_data = merged_data.sort_values('ville_population_2012', ascending=False).reset_index(drop=True)
sorted_data
```

```python
selected_data = sorted_data.head(50)
selected_data
```

```python
resort_data = selected_data.sort_values('Taux Objectif', ascending=False).reset_index(drop=True)
resort_data.index += 1
resort_data
```
